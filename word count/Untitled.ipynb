```python
import os
import matplotlib.pyplot as plt
from collections import Counter

def count_words_in_file(file_path):
    """Count the number of words in a text file."""
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
        words = text.split()
        return len(words)

def count_words_in_directory(directory):
    """Count words in all text files in the specified directory."""
    word_counts = []
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            file_path = os.path.join(directory, filename)
            word_count = count_words_in_file(file_path)
            word_counts.append(word_count)
    return word_counts

def categorize_word_counts(word_counts, bin_size=100):
    """Categorize word counts into bins."""
    max_count = max(word_counts)
    bins = range(0, max_count + bin_size, bin_size)
    bin_counts = Counter(((wc // bin_size) * bin_size for wc in word_counts))
    bin_labels = [f"{bin_size * i} - {bin_size * (i + 1) - 1}" for i in range(len(bins) - 1)]
    bin_counts_list = [bin_counts[bin_size * i] for i in range(len(bins) - 1)]
    return bin_labels, bin_counts_list

def plot_word_count_statistics(bin_labels, bin_counts):
    """Plot word count statistics as a bar chart."""
    plt.figure(figsize=(12, 6))
    plt.bar(bin_labels, bin_counts, color='skyblue')
    plt.xlabel('Word Count Ranges')
    plt.ylabel('Number of Files')
    plt.title('Distribution of Word Counts in Text Files')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Replace 'your_directory_path' with the path to your directory containing the text files
directory_path = './finished_output/'
bin_size = 100  # Set the bin size for categorizing word counts

word_counts = count_words_in_directory(directory_path)
bin_labels, bin_counts = categorize_word_counts(word_counts, bin_size)
plot_word_count_statistics(bin_labels, bin_counts)
```
